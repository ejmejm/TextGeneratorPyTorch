```python
import pickle
import pandas as pd
from collections import Counter
```

```python
with open('NYTArticles.pkl', 'rb') as f:
    headlines = pickle.load(f)
```

```python
# df.to_csv('NYTHeadlines.csv')
```

```python
df = pd.DataFrame(headlines, columns=['timestamp', 'headline', 'keywords'])
```

```python
keyword_cnt = Counter()
```

```python
for kw_list in df['keywords'].values:
    for keyword in kw_list:
        keyword_cnt[keyword] += 1
```

```python
keyword_cnt.most_common(1000)
```

```python
top_x = 1000

top_keys_decomp = {}

for key_pair in keyword_cnt.most_common(top_x):
    key = key_pair[0]
    decomp_keys = key.replace(',', '').replace('(', '').replace(')', '').split(' ')
    for sub_key in decomp_keys:
        if sub_key in top_keys_decomp:
            top_keys_decomp[sub_key] += key_pair[1]
        else:
            top_keys_decomp[sub_key] = key_pair[1]
    top_keys_decomp[key] = key_pair[1]

# del top_keys_decomp['']
```
